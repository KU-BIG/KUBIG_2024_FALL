# 타이타닉 데이터 분석

- Survived: 생존 여부 (0: 생존하지 못한 경우, 1: 생존한 경우)
- Pclass: 티켓 등급 (1: 1등석, 2: 2등석, 3: 3등석)
- Sex: 승객 성별
- Age: 승객 나이
- Sibsp: 함께 탑승한 형제/자매 또는 배우자의 수
- Parch: 함께 탑승한 부모/자녀의 수
- Fare: 운임 요금
- Embarked: 탑승 항구 (C: Cherbourg, Q: Queenstown, S: Southampton)
- Class : pclass와 동일
- Who : 남자,여자,아이
- deck : 데크 번호(알파벳 + 숫자 혼용)
- Embark_town : 탑승항구 이름
- adult_male : 성인 남자 여부
- alive : 생존여부(no, yes)
- alone : 혼자 탑승 여부

# Q3: `class` 별 `fare`에 대한 상자그림(boxplot)을 그려봅시다.

`class`에 대해 `groupby`를 한 후, 상자그림을 그려보세요

```python
# x축을 class, y축을 fare, dataset을 titanic으로 지정한 후 상자 그림을 그리는 함수 boxplot으로 그리기
# 최댓값, 최솟값, 중앙값, 사분위수 등 확인 가능, 이상치 표시
sns.boxplot(x='class', y='fare', data=titanic)
```

```python
# 산점도를 그리는 함수 scatterplot 이용
# x축은 탑승 항구, y축은 연령 분포를 나타냄.
sns.scatterplot(x='embark_town', y='age', data=titanic)
```
