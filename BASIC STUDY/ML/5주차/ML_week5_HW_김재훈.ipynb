```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import make_regression
```

Q1

```python
x, y = make_regression(n_samples=1000, n_features=10, noise = 0.1, random_state=123)
data = pd.DataFrame(x, columns=[f'x{i + 1}' for i in range(x.shape[1])])
data['y'] = y
print(data)
```

Q2

```python
#
data.hist(bins=50, figsize=(20,15))
plt.show()

# Scatterplot Matrix
sns.pairplot(data, diag_kind='hist')
plt.show()

#Correlationplot Matrix
plt.figure(figsize = (20, 10))
corr = data.corr()
sns.heatmap(corr, annot = True, cmap = "Greens")

#
data.describe()
```

Q3

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
data_scaled = pd.DataFrame(data_scaled)
print(data_scaled)
```

#Q4

```python
pca = PCA()
data_pca = pca.fit_transform(data_scaled)
var = pca.explained_variance_ratio_
data_pca = pd.DataFrame(data_pca, columns=[f'PC{i + 1}' for i in range(data_pca.shape[1])])
print(data_pca)


#Scree Plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(var) + 1), var, marker='o')
plt.xlabel("PC")
plt.ylabel("Explaned Variance Ratio")
plt.title("Screeplot")
plt.show()

#PC1, PC2 Scatter Plot
plt.figure(figsize=(20,15))
plt.scatter(data_pca['PC1'], data_pca['PC2'])
plt.xlabel('PC1', fontsize = 15)
plt.ylabel('PC2', fontsize = 15)
plt.show()
```

Q5

```python
y = data_pca['PC11']
predictors = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10']
target = ['PC11']
x = data_pca[predictors]
y = data_pca[target]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 42)

```

Q6

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

mse_train = []

for i in range(1, x_train.shape[1] + 1):
    model = LinearRegression()
    x_train_pc = x_train.iloc[:, :i]
    x_test_pc = x_test.iloc[:, :i]
    model.fit(x_train_pc, y_train)
    y_train_pred = model.predict(x_train_pc)
    mse = mean_squared_error(y_train, y_train_pred)
    mse_train.append(mse)
print(mse_train)
       
  




```
