```python
from peft import PeftModel, PeftConfig
```

```python
peft_model_id = "beomi/qlora-koalpaca-polyglot-12.8b-50step"
config = PeftConfig.from_pretrained(peft_model_id)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path, quantization_config=bnb_config, device_map={"":0})
model = PeftModel.from_pretrained(model, peft_model_id)
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)

model.eval()
```

# **Experiment 1 - 길이 고정, 파라미터 조절**

- `temperature`: 생성 텍스트의 다양성 조절.
- `top_k`: 상위 `k`개의 단어만 선택.
- `top_p`: 누적 확률이 `p` 이하가 될 때까지 단어 선택.
- `repetition_penalty`: 반복 단어 억제.

```python
# 경고 메시지 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 불필요한 메시지 방지를 위해 설정
pad_token_id = tokenizer.eos_token_id
```

## **1) 생성될 텍스트의 다양성 실험: `temperature` 조절**

```python
prompt = '인간처럼 생각하고, 행동하는 \'지능\'을 통해 인류가 이제까지 풀지 못했던'
question = "건강하게 살기 위한 세 가지 방법은?"
max_l = 64
```

```python
# 파라미터 조합 리스트
experiment_params = [
    {'temperature': 0.1, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.2, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.3, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.4, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.5, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.6, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.6, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.7, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.8, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 0.9, 'top_k': 50, 'top_p': 0.9},
    {'temperature': 1.0, 'top_k': 50, 'top_p': 0.9},
]
```

```python
with torch.no_grad():
    tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)

    for i, params in enumerate(experiment_params):
        print(f"Experiment {i+1} with params: {params}\n")

        gen_tokens = model.generate(
            tokens,
            do_sample = True,
            max_length = max_l,
            attention_mask = tokens.ne(pad_token_id),
            pad_token_id = pad_token_id,
            **params
        )
        generated = tokenizer.batch_decode(gen_tokens, skip_special_tokens=True)[0]
        print(f"Generated Text for Prompt: {generated}\n")

        # 질문에 대한 답변 생성
        q = f"### 질문: {question}\n\n### 답변:"
        question_tokens = tokenizer.encode(q, return_tensors = 'pt').to(device='cuda', non_blocking=True)
        question_mask = question_tokens.ne(pad_token_id)
        gened = model.generate(
            question_tokens,
            attention_mask=question_mask,
            pad_token_id=pad_token_id,
            max_new_tokens=50,
            early_stopping=True,
            do_sample=True,
            eos_token_id=2,
            **params
        )
        answer = tokenizer.decode(gened[0], skip_special_tokens=True)
        print(f"Generated Answer for Question: {answer}\n")

        print("-"*50 + "\n\n\n\n")
```

`temperature` 값을 높게 설정할수록 확실히 답변의 다양성이나 추상성 등이 높아지는 것으로 보인다. 값이 낮을 때는 질문을 그대로 반복하기도 하고 혹은 '건강한 운동', '건강한 음식', '건강한 마음'과 같이 반복적이면서도 당연한(?) 말을 한다. 0.3 정도부터 어느 정도 말이 되는 답변을 내놓는데, 0.7 정도 값 까지는 추상적인 것보다는 어떤 구체적인 행동을 답변으로 내놓는다. 그러다가 점점 더 추상적인 답변을 하거나 혹은 ~해야 ~한다. 와 같은 일종의 논증을 하거나 답변과 그것에 대한 근거를 말하기도 한다. 그리고 높은 값일 때 더 긴 답변을 하고 싶어 하는 것 같다. 물론 갑자기 뜬금 없는 말을 하기도 한다.

이 질문의 경우에는 1.2 정도인 경우가 가장 정확하면서도 반복이 없는 답변이다. (물론 히어로즈가 빠졌다..)

## **정리**

- temperature: 0~1 사이의 값으로 조절하고, 1에 가까울수록 더욱 다양하고 추상적인 텍스트를 생성해낸다. 0.8 정도가 적당한 것 같다.

- top_k: 프롬프트나 질문과 연관된 것 중 어느 정도 범위로 관련된 것들을 생성해낼지 결정하는 파라미터인 것 같다. 값이 높을수록 마치 MBTI가 N인 사람처럼 더 넓은 범위에서의 연관 키워드들을 기반으로 텍스트를 생성한다. 50 정도가 적당?

- top_p: 아직 정확하게 어떤 기능인지는 이해하지 못함

- repetition_penalty: 프롬프트나 질문과 연관된 키워드와 내용을 기반으로 생성될 내용을 얼마나 반복할지 조절하는 파라미터. 1이거나 설정하지 않으면 많이 반복하는 것 같고, 2에 가까울수록 반복이 확연히 방지되지만 그만큼 짧은 텍스트를 생성해내려고 하는 것 같다. 1.2 정도가 적당?

### **0) 우선 길이 유지**

```python
prompt = '시간에 시작점이 있다면 그 시작점 이전의 시간은 어떤 의미를 가질까? 시간은 우주의 본질적인 것인가, '
question = "태양계에는 어떤 행성들이 있고 각각 어떤 특성을 가지고 있을까?"
```

```python
max_l = 64
max_new_token = 50
```

```python
# 파라미터 조합 리스트
experiment_params = [
    {'temperature': 0.8, 'top_k': 50, 'top_p': 0.9, 'repetition_penalty': 1.2},
    {'temperature': 0.8, 'top_k': 60, 'top_p': 0.9, 'repetition_penalty': 1.2},
    {'temperature': 0.8, 'top_k': 70, 'top_p': 0.9, 'repetition_penalty': 1.2},
    {'temperature': 0.8, 'top_k': 80, 'top_p': 0.9, 'repetition_penalty': 1.2},
    {'temperature': 0.8, 'top_k': 90, 'top_p': 0.9, 'repetition_penalty': 1.2},
    {'temperature': 0.8, 'top_k': 100, 'top_p': 0.9, 'repetition_penalty': 1.2},
]
```

```python
with torch.no_grad():
    tokens = tokenizer.encode(prompt, return_tensors='pt').to(device='cuda', non_blocking=True)

    for i, params in enumerate(experiment_params):
        print(f"Experiment {i+1} with params: {params}\n")

        gen_tokens = model.generate(
            tokens,
            do_sample = True,
            max_length = max_l,
            attention_mask = tokens.ne(pad_token_id),
            pad_token_id = pad_token_id,
            **params
        )
        generated = tokenizer.batch_decode(gen_tokens, skip_special_tokens=True)[0]
        print(f"Generated Text for Prompt: {generated}\n")

        # 질문에 대한 답변 생성
        q = f"### 질문: {question}\n\n### 답변:"
        question_tokens = tokenizer.encode(q, return_tensors = 'pt').to(device='cuda', non_blocking=True)
        question_mask = question_tokens.ne(pad_token_id)
        gened = model.generate(
            question_tokens,
            attention_mask=question_mask,
            pad_token_id=pad_token_id,
            max_new_tokens = max_new_token,
            early_stopping=True,
            do_sample=True,
            eos_token_id=2,
            **params
        )
        answer = tokenizer.decode(gened[0], skip_special_tokens=True)
        print(f"Generated Answer for Question: {answer}\n")

        print("-"*50 + "\n\n\n\n")
```
