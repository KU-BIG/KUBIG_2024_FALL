```python
import pandas as pd
import matplotlib.pyplot as plt

# 공통 함수 정의: 총 공급량, 총 사용량, 유수율 및 이동평균 유수율 계산 및 시각화
def visualize_network(df, network_name, supply_nodes, demand_nodes, rolling_window=3):
    # 공급량 및 사용량 계산
    df['_Qin'] = df[supply_nodes].sum(axis=1)
    df['_Qout'] = df[demand_nodes].sum(axis=1)
    df['_Qratio'] = df['_Qout'] / df['_Qin']
    df['_Qratio2'] = df['_Qout'].rolling(rolling_window).mean() / df['_Qin'].rolling(rolling_window).mean()
    
    # 시각화: 총 공급량과 총 사용량
    plt.figure(figsize=(10, 6))
    plt.plot(df['timestamp'], df['_Qin'], label='총 공급량', color='blue')
    plt.plot(df['timestamp'], df['_Qout'], label='총 사용량', color='orange')
    plt.title(f'{network_name} 관망 - 총 공급량과 총 사용량')
    plt.xlabel('시간')
    plt.ylabel('유량')
    plt.legend()
    plt.grid()
    plt.show()
    
    # 시각화: 유수율 및 이동평균 유수율
    plt.figure(figsize=(10, 6))
    plt.plot(df['timestamp'], df['_Qratio'], label='유수율', color='blue')
    plt.plot(df['timestamp'], df['_Qratio2'], label='이동평균 유수율', color='red')
    plt.title(f'{network_name} 관망 - 유수율 및 이동평균 유수율')
    plt.xlabel('시간')
    plt.ylabel('유수율')
    plt.legend()
    plt.grid()
    plt.show()

# C와 D 관망 타임스탬프 변환 함수
def convert_timestamp(timestamp_col):
    # "T"를 0으로 변환
    timestamp_col = timestamp_col.replace("T", "T-0")
    # "T-숫자"에서 숫자만 추출하고 음수 변환
    return timestamp_col.str.extract(r'T-(\d+)').astype(int) * -1
```

```python
# 관망 A 데이터 로드
df_a = pd.read_csv('./train/TRAIN_A.csv')

# 타임스탬프 변환
df_a['timestamp'] = pd.to_datetime(df_a['timestamp'], format='%y/%m/%d %H:%M')

# 관망 A 시각화
visualize_network(df_a, "A 관망", supply_nodes=['Q1', 'Q2', 'Q3', 'Q4'], demand_nodes=['Q5'])
```

```python
# 관망 B 데이터 로드
df_b = pd.read_csv('./train/TRAIN_B.csv')

# 타임스탬프 변환
df_b['timestamp'] = pd.to_datetime(df_b['timestamp'], format='%y/%m/%d %H:%M')

# 관망 B 시각화
visualize_network(df_b, "B 관망", supply_nodes=['Q1'], demand_nodes=['Q2', 'Q3', 'Q4'])
```

```python
# 관망 C 데이터 로드
df_c = pd.read_csv('./test/C/TEST_C_0000.csv')

# 타임스탬프 변환
df_c['timestamp'] = convert_timestamp(df_c['timestamp'])

# 관망 C 시각화
visualize_network(df_c, "C 관망", supply_nodes=['Q1'], demand_nodes=['Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8'])
```

```python
# 관망 D 데이터 로드
df_d = pd.read_csv('./test/D/TEST_D_0000.csv')

# 타임스탬프 변환
df_d['timestamp'] = convert_timestamp(df_d['timestamp'])

# 관망 D 시각화
visualize_network(df_d, "D 관망", supply_nodes=['Q1'], demand_nodes=['Q2', 'Q3', 'Q4', 'Q5'])
```
