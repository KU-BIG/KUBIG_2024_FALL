```python
import pandas as pd

# 주별 판매량 저장할 리스트
weekly_sales_list = []

# 각 병합된 데이터프레임에 대해 주별로 그룹화
for i, merged_df in enumerate(dfs):
    # 1. '판매일'을 datetime 형식으로 변환 (이 과정이 누락되어 발생한 에러를 해결)
    merged_df['판매일'] = pd.to_datetime(merged_df['판매일'])

    # 2. '판매일'을 기준으로 주별로 그룹화하고 '판매수량'을 합산
    weekly_sales = merged_df.groupby(
        pd.Grouper(key='판매일', freq='W')
    )['판매수량'].sum().reset_index()

    # 3. 주별 판매량을 리스트에 저장
    weekly_sales_list.append(weekly_sales)

    # 4. 주별 판매량 출력 확인
    print(f"주별 판매량 {i+1}:")
    print(weekly_sales.head())

# 이제 weekly_sales_list에 각 데이터프레임에 대한 주별 판매량이 저장되었습니다.
```

```python
import pandas as pd


# 1. price_df의 '년월'을 문자열(YYYY-MM)로 변환
price_df['년월'] = price_df['년월'].astype(str)

# 각 주별 판매량 데이터프레임에 대해 년월을 기준으로 병합
for i, weekly_sales in enumerate(weekly_sales_list):
    # 2. '판매일'을 datetime 형식으로 변환하고 '년월' 정보 추출 (YYYY-MM)
    weekly_sales['판매일'] = pd.to_datetime(weekly_sales['판매일'])  # 날짜 형식으로 변환
    weekly_sales['년월'] = weekly_sales['판매일'].dt.to_period('M').astype(str)  # 년월 정보 추출

    # 3. weekly_sales와 price_df를 '년월'을 기준으로 병합 (물가지수 추가)
    weekly_sales_merged = pd.merge(weekly_sales, price_df[['년월', '물가지수']], on='년월', how='left')

    # 병합된 결과 확인
    print(f"병합된 주별 판매량 데이터프레임 {i+1}:")
    print(weekly_sales_merged.head())

    # 병합된 데이터프레임을 원래 리스트에 반영
    weekly_sales_list[i] = weekly_sales_merged
```

```python
# weekly_sales_list에 있는 모든 데이터프레임에 대해 lag 피처 생성
for i, weekly_sales in enumerate(weekly_sales_list):
    for lag in range(1, 4):
        # 각 데이터프레임에 대해 lag_1, lag_2, lag_3 피처 생성
        weekly_sales[f'lag_{lag}'] = weekly_sales['판매수량'].shift(lag)

    # 결과 확인
    print(f"weekly_sales_{i+1}에 lag 피처 추가:")
    print(weekly_sales.head())

    # 업데이트된 데이터프레임을 리스트에 다시 반영
    weekly_sales_list[i] = weekly_sales
```

```python
# weekly_sales_list에 있는 모든 데이터프레임에 대해 lag 피처의 결측치를 최솟값으로 대체
for i, weekly_sales in enumerate(weekly_sales_list):
    for lag in range(1, 4):
        # 결측치를 해당 컬럼의 최솟값으로 대체
        min_value = weekly_sales[f'lag_{lag}'].min()  # 최솟값 계산
        weekly_sales[f'lag_{lag}'].fillna(min_value, inplace=True)  # 결측치 대체

    # 결과 확인
    print(f"weekly_sales_{i+1}의 lag 피처 결측치 대체 완료:")
    print(weekly_sales.head())

    # 업데이트된 데이터프레임을 리스트에 다시 반영
    weekly_sales_list[i] = weekly_sales
```
