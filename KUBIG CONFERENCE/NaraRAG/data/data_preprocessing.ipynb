```python
questioner = []
answerer = []
for idx, row in df.iterrows():
    try:
        q_name = row['questioner_name']+' '+row['questioner_affiliation']+' '+row['questioner_position']
    except:
        if type(row['questioner_affiliation'])==float:
            if type(row['questioner_position'])==float:
                q_name = row['questioner_name']
            else:
                q_name = row['questioner_name']+' '+row['questioner_position']
        else:
            q_name = row['questioner_name']+' '+row['questioner_affiliation']
            
    
    try:
        a_name = row['answerer_name']+' '+row['answerer_affiliation']+' '+row['answerer_position']
    except:
        if type(row['answerer_affiliation'])==float:
            if type(row['answerer_position'])==float:
                a_name = row['answerer_name']
            else: a_name = row['answerer_name']+' '+row['answerer_position']
        else: a_name = row['answerer_name']+' '+row['answerer_affiliation']
            
    
    questioner.append(q_name)
    answerer.append(a_name)
```

```python
law_length = df['법안'].apply(lambda x: len(x) if type(x)==str else 0)
agenda_length = df['안건'].apply(lambda x: len(x) if type(x)==str else 0)
```

```python
law_only = pd.concat([df['법안'], pd.Series(law_length, name='length')], axis=1)
agenda_only = pd.concat([df['안건'], pd.Series(agenda_length, name='length')], axis=1)
```

```python
len(law_only)
```

```python
comma_check = law_only[law_only['법안'].apply(lambda x: ',' in str(x))]
comma_check.sort_values('length').iloc[100:150]
```

```python
def comma_split(x):
    if type(x) == str:
        return x.split(',')
    else:
        return None
law_only['list'] = law_only['법안'].apply(lambda x: comma_split(x))
```

```python
law_only['list_length'] = law_only['list'].apply(lambda x: len(x) if type(x)==list else 0)
```

```python
law_only.sort_values('list_length', ascending=False).iloc[50:160]
```

```python
def update_law(x):
    if isinstance(x['list'], list) and len(x['list']) >= 3:
        return x['list'][-1]
    return x['법안']

law_only['법안_new'] = law_only.apply(update_law, axis=1)
```

```python
law_only.sort_values('list_length', ascending=False).iloc[50:160]
```

```python
law_only['length_new']=law_only['법안_new'].apply(lambda x: len(x) if type(x)==str else 0)
law_only.sort_values('length_new', ascending=False).head(50)
```
