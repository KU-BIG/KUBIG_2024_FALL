```python
import asyncio
import time
from langchain_chroma import Chroma
from langchain_upstage import UpstageEmbeddings

DB_PATH = './chroma_db_3'

class ChromaParallel(Chroma):

    async def afrom_documents(documents, embedding, num_workers=3):
        db = Chroma(embedding_function=embedding, persist_directory=DB_PATH)
        # create list of num_workers empty lists
        doc_groups = [[] for _ in range(num_workers)]

        for i in range(len(documents)):
            doc_groups[i % num_workers].append(documents[i])

        tasks = [db.aadd_documents(group) for group in doc_groups]
        await asyncio.gather(*tasks)
        return db

now = time.time()

# 3. Embed & indexing
loop = asyncio.get_event_loop()
vectorstore = await ChromaParallel.afrom_documents(
    documents=docs,
    embedding= UpstageEmbeddings(model="solar-embedding-1-large"),
    num_workers=5
)

#retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
```

```python
from langchain.chains.query_constructor.base import (
    StructuredQueryOutputParser,
    get_query_constructor_prompt,
)

# 문서 내용 설명과 메타데이터 필드 정보를 사용하여 쿼리 생성기 프롬프트를 가져옵니다.
prompt = get_query_constructor_prompt(
    "Brief summary of a cosmetic product",  # 문서 내용 설명
    metadata_field_info,  # 메타데이터 필드 정보
)

# StructuredQueryOutputParser 를 생성
output_parser = StructuredQueryOutputParser.from_components()

# query_constructor chain 을 생성
query_constructor = prompt | llm | output_parser
```

```python
query_output = query_constructor.invoke(
    {
        # 쿼리 생성기를 호출하여 주어진 질문에 대한 쿼리를 생성합니다.
        "query": "조윤선이 질문한 발언들을 나열해주세요"
    }
)
```
