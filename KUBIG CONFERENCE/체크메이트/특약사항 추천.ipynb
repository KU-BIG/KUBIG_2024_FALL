```python
import numpy as np
from langchain_upstage import UpstageEmbeddings

def get_embedding(text):
    embeddings = UpstageEmbeddings(api_key=os.environ["API_KEY"], model="solar-embedding-1-large-passage")
    response = embeddings.embed_documents([text])  # Upstage는 리스트 형식 입력 요구
    return np.array(response[0]) 
```

```python
embedded_agreements = [get_embedding(text) for text in special_agreements]
```

```python
clause = "임대인은 세금 이자 체납이 없음을 고지하며, 임차인이 세금 체납 내역을 확인하는 것에 협조한다. 만일 세금 체납이 확인된다면 잔금일 이전까지 체납액을 모두 상환하며, 이를 어길 시에는 계약은 무효로 한다."
embedded_clause = get_embedding(clause)
distances = [np.linalg.norm(embedded_clause - emb) for emb in embedded_agreements]

most_similar_index = np.argmin(distances)
most_similar_sentence = special_agreements[most_similar_index]

print("입력:", clause)
print("가장 유사한 특약사항:", most_similar_sentence)
print("유클리드 거리:", distances[most_similar_index])
```
