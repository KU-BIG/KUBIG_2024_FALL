```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time

# 옵션 및 드라이버 설정
options = Options()
options.add_argument("--no-sandbox")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 네이버 지도 페이지로 이동
url = "https://m.place.naver.com/place/list?query=강릉시%20음식점&level=top"
driver.get(url)

# 페이지 로드 완료까지 대기
driver.implicitly_wait(10)

# 스크롤을 수행하여 모든 결과를 불러오기
scroll_pause_time = 1  # 스크롤 후 대기 시간
last_height = driver.execute_script("return document.body.scrollHeight")
max_attempts = 50
attempts = 0

while attempts < max_attempts:
    # 페이지 끝으로 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)

    # 새로운 높이 확인
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    if new_height == last_height:
        attempts += 1
        print(f"더 이상 로드되는 항목이 없습니다. {attempts}회 확인 중.")
    else:
        last_height = new_height
        attempts = 0  # 새로운 항목이 로드된 경우 시도 횟수 초기화

    # 시도 횟수가 제한을 넘으면 루프 종료
    if attempts >= max_attempts:
        print("모든 데이터를 로드한 것으로 보입니다.")
        break

# 현재 페이지에서 데이터 수집
restaurants = []
elements = driver.find_elements(By.CSS_SELECTOR, "li.VLTHu")
print(f"발견된 음식점 개수: {len(elements)}")

for element in elements:
    try:
        # 음식점 이름
        name = WebDriverWait(element, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "span.YwYLL"))
        ).text
        
        # 카테고리 (음식 종류)
        try:
            category = element.find_element(By.CSS_SELECTOR, "span.YzBgS").text
        except:
            category = None

        # 전화번호
        try:
            phone_number = element.find_element(By.CSS_SELECTOR, "a._T0lO").get_attribute("href").replace("tel:", "")
        except:
            phone_number = None

        # 상세 페이지 URL
        detail_url = element.find_element(By.CSS_SELECTOR, "a.P7gyV").get_attribute("href")

        # 수집한 정보 저장
        restaurants.append({
            '이름': name,
            '카테고리': category,
            '전화번호': phone_number,
            '상세 URL': detail_url,
        })
        print(f"수집 완료 - 이름: {name}, 카테고리: {category}, 전화번호: {phone_number}, URL: {detail_url}")

    except Exception as e:
        print(f"요소를 찾는 중 오류 발생: {e}")
        continue

# DataFrame으로 저장
df = pd.DataFrame(restaurants)
print("수집한 데이터 요약:")
print(df.head())
print(f"총 수집된 음식점 수: {len(df)}")

# 드라이버 종료
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time


def review_scraper_v2(df):
    # 옵션 설정
    options = Options()
    options.add_argument("--no-sandbox")

    # 크롬 드라이버 설정 (드라이버를 한 번만 실행)
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    # 수집할 기본 상세 URL 목록
    base_urls = list(df["상세 URL"])

    # 수집한 데이터를 저장할 리스트
    data = []

    for base_url in base_urls:
        try:
            # 기본 URL 접근
            driver.get(base_url)
            driver.implicitly_wait(10)
            
            # 기본 정보 수집
            try:
                # 방문자 리뷰 수
                visitor_reviews = driver.find_element(By.CSS_SELECTOR, 'span.PXMot a[href*="review/visitor"]').text
            except:
                visitor_reviews = None

            try:
                # 블로그 리뷰 수
                blog_reviews = driver.find_element(By.CSS_SELECTOR, 'span.PXMot a[href*="review/ugc"]').text
            except:
                blog_reviews = None

            try:
                # 별점
                rating = driver.find_element(By.CSS_SELECTOR, 'span.PXMot.LXIwF').text
            except:
                rating = None

            try:
                # 세부 주소
                address = driver.find_element(By.CSS_SELECTOR, 'span.LDgIH').text
            except:
                address = None

            # 방문자 리뷰 페이지 URL 구성 및 이동
            restaurant_id = base_url.split('/')[-1].split("?")[0]  # URL에서 restaurant ID 추출
            review_url = f"https://m.place.naver.com/restaurant/{restaurant_id}/review/visitor?entry=ple"
            driver.get(review_url)
            driver.implicitly_wait(10)

            # 리뷰 텍스트 수집
            reviews = []
            review_counter = 0

            # 페이지 스크롤 및 리뷰 '더보기' 버튼 처리
            while review_counter < 20:
                # 개별 리뷰에서 '더보기' 버튼을 클릭해 전체 리뷰 텍스트를 불러옴
                review_elements = driver.find_elements(By.CSS_SELECTOR, 'div.pui__vn15t2')  # 리뷰 텍스트 요소 선택
                for element in review_elements:
                    try:
                        # 리뷰의 '더보기' 버튼 클릭
                        more_button = element.find_element(By.CSS_SELECTOR, 'a.pui__wFzIYl')
                        more_button.click()
                        time.sleep(0.2)
                    except:
                        pass
                    # 전체 리뷰 텍스트 저장
                    reviews.append(element.text)
                    review_counter += 1

                # 페이지 하단의 '더보기' 버튼 클릭하여 추가 리뷰 로드
                try:
                    load_more_button = driver.find_element(By.XPATH, '//*[@id="app-root"]/div/div/div/div[6]/div[2]/div[3]/div[2]/div/a')
                    load_more_button.click()
                    time.sleep(0.4)
                except:
                    print("모든 리뷰 로드 완료")
                    break

            # 수집한 정보 저장
            data.append({
                "URL": base_url,
                "방문자 리뷰 수": visitor_reviews,
                "블로그 리뷰 수": blog_reviews,
                "평점": rating,
                "세부 주소": address,
                "리뷰 텍스트": reviews
            })

            print(f"{base_url} 데이터 수집 완료")
            print({
                "URL": base_url,
                "방문자 리뷰 수": visitor_reviews,
                "블로그 리뷰 수": blog_reviews,
                "평점": rating,
                "세부 주소": address,
                "리뷰 텍스트": reviews
            })

        except Exception as e:
            print(f"오류 발생 - URL: {base_url}, 오류: {e}")
            continue

    # 모든 URL 처리 후 드라이버 종료
    driver.quit()

    return pd.DataFrame(data)
```
