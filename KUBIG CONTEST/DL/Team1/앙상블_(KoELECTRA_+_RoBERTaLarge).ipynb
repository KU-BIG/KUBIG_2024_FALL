```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH =  '/content/drive/MyDrive'
```

```python
output_prob_1 = pd.read_csv(os.path.join(PATH, 'output_prob_df_electra.csv'), encoding='utf-8')
output_prob_2 = pd.read_csv(os.path.join(PATH, 'output_prob_df_robert.csv'), encoding='utf-8')
```

```python
electra_list = output_prob_1.values
print(electra_list)
```

```python
robert_list = output_prob_2.values
print(robert_list)
```

```python
# 각각의 가중치 0.4, 0.6
koelectra_probs_w1 = 0.4
klue_roberta_probs_w2 = 0.6

average_probs = [koelectra_probs_w1 * p1 + klue_roberta_probs_w2 * p2 for p1, p2 in zip(electra_list, robert_list)]

print(average_probs)
```

```python
pred_answer = []
for prob in average_probs:
  pred_answer.append(np.argmax(prob, axis = -1))

print(pred_answer)
```

```python
def num_to_label(label):
    label_dict = {0: "entailment", 1: "contradiction", 2: "neutral"}
    str_label = []

    for i, v in enumerate(label):
        str_label.append([i,label_dict[v]])

    return str_label

answer = num_to_label(pred_answer)
print(answer)
```

```python
df = pd.DataFrame(answer, columns=['index', 'label'])

print(df)
```

```python
df.to_csv('./result/submission_ensemble.csv', index=False)
```
